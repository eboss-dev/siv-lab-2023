## Objective: 

Learn how to apply various filters to an image using OpenCV, NumPy, and Matplotlib.

```python
#read image

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('imgs_lab/lena_color.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Implement functions to apply various filters (gaussian blur, median blur and sobel edge detection) to the image

- https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

- https://www.geeksforgeeks.org/python-program-to-detect-the-edges-of-an-image-using-opencv-sobel-edge-detection/


```python
def apply_gaussian_blur(image, kernel_size=(5, 5)):
    blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
    return blurred_image

def apply_median_blur(image, kernel_size=5):
    blurred_image = cv2.medianBlur(image, kernel_size)
    return blurred_image

def apply_sobel_edge_detection(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel_image = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=5)
    return sobel_image
```

Display the original image and the filtered images using Matplotlib

```python
plt.figure(figsize=(15, 5))
plt.subplot(141), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image'), plt.axis('off')

gaussian_blurred = apply_gaussian_blur(image)
plt.subplot(142), plt.imshow(cv2.cvtColor(gaussian_blurred, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blur'), plt.axis('off')

median_blurred = apply_median_blur(image)
plt.subplot(143), plt.imshow(cv2.cvtColor(median_blurred, cv2.COLOR_BGR2RGB))
plt.title('Median Blur'), plt.axis('off')

edge_image = apply_sobel_edge_detection(image)
plt.subplot(144), plt.imshow(edge_image)
plt.title('Edges'), plt.axis('off')

plt.show()
```
